```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
import os
import math
```

```python
compare_1 = '/home/van/research/EDoF-CNN/img_res/EDOF_CNN_pack_43'
compare_2 = '/home/van/research/EDoF-CNN/img_res/EDOF_CNN_fast_4'
PRED_NAME = 'PRED_'
GT_NAME = 'GT_'
IMG_NAME = ['0','1','2']
```

```python
cur_fold = '0'
row_images = []
cur_fold_compare_1 = os.path.join(compare_1,cur_fold)
cur_fold_compare_2 = os.path.join(compare_2,cur_fold)
for cur_img_name in IMG_NAME:
    full_img_name = PRED_NAME + cur_img_name + '.png'
    full_gt_img_name = GT_NAME + cur_img_name + '.png'
    compare_1_image_path = os.path.join(cur_fold_compare_1,full_img_name)
    compare_2_image_path = os.path.join(cur_fold_compare_2,full_img_name)
    gt_image_path = os.path.join(cur_fold_compare_1, full_gt_img_name)
    compare_1_image = cv2.imread(compare_1_image_path)
    compare_2_image = cv2.imread(compare_2_image_path)
    gt_image = cv2.imread(gt_image_path)
    row = np.array([gt_image,compare_1_image,compare_2_image])
    row_images.append(row)
```

```python
w = 10
h = 10
fig = plt.figure(figsize=(8, 8))
columns = 3
rows = 3
for i in range(1, columns*rows +1):
    param_row = math.ceil(i/3)
    param_col = (i+2)%3
    print("param_row: " + str(param_row) + " param_col: " + str(param_col))
    img = row_images[param_row][param_col]
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
plt.show()
```
